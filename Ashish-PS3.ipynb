```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Question 1

```python
users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head()
age_mean = users.groupby('occupation')['age'].mean()
occupation_wise_count = users['occupation'].value_counts()
male_occupation = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)['M'] / occupation_wise_count * 100
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
occupation_gender_mean_age = users.groupby(['occupation', 'gender'])['age'].mean()
occupation_gender_count = users.groupby('occupation')['gender'].count()
occupation_percent_gender = (users.groupby(['occupation', 'gender'])['gender'].count() / occupation_gender_count * 100)

print("\nAns Step 4: \n" + str(age_mean))
print("\nAns Step 5: \n" + str(male_occupation))
print("\nAns Step 6: \n" + str(min_max_age))
print("\nAns Step 7: \n" + str(occupation_gender_mean_age))
print("\nAns Step 8: \n" + str(occupation_percent_gender))
```

### Question 6

```python
marriage_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
marriage_data = marriage_data[(marriage_data.Year >= 1867) & (marriage_data.Year <= 2014)]
marriage_data['Marriages_per_capita'] = marriage_data.Marriages / marriage_data.Population
marriage_data['Divorces_per_capita'] = marriage_data.Divorces / marriage_data.Population

print(type(marriage_data.Marriages_per_capita))

plt.plot(marriage_data.Year, marriage_data.Marriages_per_capita, label='Marriages')
plt.plot(marriage_data.Year, marriage_data.Divorces_per_capita, label='Divorces')
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()


plt.show()
```

### Question 9

```python
data = pd.read_csv('roman-emperor-reigns.csv')
total_emperors = len(data)
assassinated_emperors = len(data[data['Cause_of_Death'] == 'Assassinated'])
percentage_assassinated = (assassinated_emperors / total_emperors) * 100
percentage_survived = 100 - percentage_assassinated


labels = ['Assassinated', 'Survived']
sizes = [percentage_assassinated, percentage_survived]
colors = ['red', 'green']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Fraction of All Roman Emperors Assassinated')
plt.axis('equal')


plt.show()
```
